```python
# Plot 100 Wiener processes:
from matplotlib import pyplot as plt
import torch as th

import numpy as np
import matplotlib.pyplot as plt

def generate_wiener_process(T, N, M):
    dt = T/N
    dW = np.sqrt(dt) * np.random.normal(0, 1, (N, M))
    W = np.cumsum(dW, axis=0)
    W = np.vstack((np.zeros(M), W))  # insert W_0 = 0 at the start of every path
    return W

def plot_wiener_process(T, N, M):
    W = generate_wiener_process(T, N, M)
    t = np.linspace(0.0, T, N+1)

    for m in range(M):
        plt.plot(t, W[:,m])

    plt.title(f'{M} Wiener Processes')
    plt.xlabel('$t$')
    plt.ylabel('$W_t$')
    plt.grid(True)
    plt.savefig('wiener_process.pdf')
    plt.show()

# Plot 100 Wiener processes over the time period [0,1] using 1000 discrete steps.
plot_wiener_process(1, 1000, 50)
```
